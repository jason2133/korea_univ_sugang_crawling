```python
import urllib.request as req
from bs4 import BeautifulSoup
import pandas as pd
```

```python
your_major = "STAT" # 전공과목영문코드 입력
cour_num = []
name = []
cour_time = []
cour_prof = []
```

```python
!nvidia-smi
```

```python
for number in range (100, 500):
    url = "https://infodepot.korea.ac.kr/lecture1/lecsubjectPlanView.jsp?year=2023&term=1R&grad_cd=0136&col_cd=9999&dept_cd=4653&cour_cd=" + your_major + str(number) + "&cour_cls=00&"
    html = req.urlopen(url)
    doc = BeautifulSoup(html,"html.parser")
    temp_title = doc.h3.string
    if len(temp_title) == 19:
        for number2 in range (1, 10):
            url2 = url[:-2] + str(number2) + "&"
            html = req.urlopen(url2)
            doc = BeautifulSoup(html,"html.parser")
            temp_title = doc.h3.string
            if len(temp_title) == 19:
                break
            else:
                title = temp_title.split("[")[1].split("]")[0]
                try:
                    time = doc.find(class_="tbl_view").tbody.tr.td.string.splitlines()
                    time = " ".join(time)
                    prof = doc.find(class_="tbl_view").next_sibling.next_sibling.next_sibling.next_sibling.tbody.tr.td.string
                except:
                    time = None
                    prof = None
                cour_time.append(time)
                cour_prof.append(prof)
                cour_number = your_major + str(number) + "-0" + str(number2)
                cour_num.append(cour_number)
                name.append(title)
                print(f"{cour_number} 완료")
    else:
        title = temp_title.split("[")[1].split("]")[0]
        try:
            time = doc.find(class_="tbl_view").tbody.tr.td.string.splitlines()
            time = " ".join(time)
            prof = doc.find(class_="tbl_view").next_sibling.next_sibling.next_sibling.next_sibling.tbody.tr.td.string
        except:
            time = None
            prof = None
        cour_time.append(time)
        cour_prof.append(prof)
        cour_number = your_major + str(number) + "-00"
        cour_num.append(cour_number)
        name.append(title)
        print(f"{cour_number} 완료")
```

```python
cour_table = pd.DataFrame({"학수번호" : cour_num, "강의명" : name, "시간" : cour_time, "교수명" : cour_prof})
cour_table
```

```python
cour_table_2 = cour_table.drop(cour_table[cour_table["강의명"] == " "].index)
cour_table_2.to_excel(f"23_1_{your_major}.xlsx")
```
